### ① 取出今天所有的發文

```python
import requests
from bs4 import BeautifulSoup

url = 'https://www.ettoday.net/news/news-list.htm'
r = requests.get(url)

soup = BeautifulSoup(r.text, "html5lib")

for d in soup.find(class_="part_list_2").find_all('h3'):
    print(d.find(class_="date").text, d.find_all('a')[-1].text)
```

### ② 如果想要依照類別分類，怎麼儲存會比較好？

```python
tag = []
em = soup.find(class_="part_list_2").find_all('h3')
for d in em:
#     print(d.em['class'][1])
    if not d.em['class'][1] in tag:
        tag.append(d.em['class'][1])

# for i in range(len(tag)):
t = soup.find(class_="part_list_2").find_all('h3')
for d in t:
    print(d.em['class'][1])
    print(d.em.text)
    print(d.find(class_="date").text, d.find_all('a')[-1].text)
```

### ③ 哪一個類別的文章最多

```python
from collections import Counter

tag = []

em = soup.find(class_="part_list_2").find_all('em')
for d in em:
    tag.append(d.text)

result = Counter(tag)
print(result.most_common(1))
```
