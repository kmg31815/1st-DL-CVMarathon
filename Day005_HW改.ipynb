```python
# 1. é€™å€‹ API ä¸€æ¬¡æœƒå›å‚³å¹¾ç­†è³‡æ–™ï¼Ÿæ¯ä¸€ç­†è³‡æ–™åŒ…å«å“ªäº›æ¬„ä½ï¼Ÿ

import requests
r = requests.get('https://www.dcard.tw/_api/forums/pet/posts?popular=true')
response = r.text

import json
data = json.loads(response)
# data

print("å›å‚³è³‡æ–™:",len(data),"ç­†,\n")

print(data[0].keys())
```

```python
# 2. å–å‡ºæ¯ä¸€ç­†è³‡æ–™çš„ã€Œæ¨™é¡Œã€ã€ã€Œè²¼æ–‡æ™‚é–“ã€ã€ã€Œç•™è¨€äººæ•¸ã€ã€ã€ŒæŒ‰è®šäººæ•¸ã€

for d in data:
    print("æ¨™é¡Œ:",d['title'])
    print("è²¼æ–‡æ™‚é–“:",d['updatedAt'])
    print("ç•™è¨€äººæ•¸:",d['commentCount'])
    print("æŒ‰è®šäººæ•¸:",d['likeCount'])
    print("")
```

```python
# 3. è¨ˆç®—ç†±é–€/éç†±é–€æ–‡ç« çš„ã€Œå¹³å‡ç•™è¨€äººæ•¸ã€èˆ‡ã€Œå¹³å‡æŒ‰è®šäººæ•¸ã€

commentCount = 0
likeCount = 0

for d in data:
    commentCount += d['commentCount']
    likeCount += d['likeCount']

print("ç†±é–€å¹³å‡ç•™è¨€äººæ•¸:",commentCount/len(data))
print("ç†±é–€å¹³å‡æŒ‰è®šäººæ•¸:",likeCount/len(data))

r2 = requests.get('https://www.dcard.tw/_api/forums/pet/posts?popular=false')
response2 = r2.text

data2 = json.loads(response2)

commentCount2 = 0
likeCount2 = 0

for d2 in data2:
    commentCount2 += d2['commentCount']
    likeCount2 += d2['likeCount']

print("éç†±é–€å¹³å‡ç•™è¨€äººæ•¸:",commentCount2/len(data2))
print("éç†±é–€å¹³å‡æŒ‰è®šäººæ•¸:",likeCount2/len(data2))
```
